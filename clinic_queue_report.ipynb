## **1. Mục tiêu**

Mục tiêu của bài tập lớn này là xây dựng và mô phỏng một mô hình phòng khám nha khoa nhiều giai đoạn (multi-stage dental clinic queuing system) nhằm xác định các yếu tố cơ sở vật chất ảnh hưởng đến hiệu năng hoạt động của hệ thống phòng khám. Hệ thống được thiết kế để mô phỏng quy trình bệnh nhân từ lúc đăng ký khám bệnh (Check-in) đến khi hoàn tất điều trị và rời phòng khám (Check-out), đi qua nhiều phòng xử lý trung gian như phòng khám (Treatment Room), phòng chụp X-quang (X-ray Room), phòng phẫu thuật (Surgery Room), và phòng hồi phục (Recovery Room).

### **1.1. Phạm vi nghiên cứu**

Trong nghiên cứu này, hệ thống phòng khám được xem xét dựa trên **ba yếu tố cơ sở vật chất chính**, có ảnh hưởng trực tiếp đến hiệu năng vận hành:

a. **Số server trong mỗi hàng đợi (Number of servers per queue)**  
   - Tương ứng với số lượng **ghế nha sĩ hoặc nhân viên phục vụ song song** ở mỗi khu vực.  
   - Ảnh hưởng trực tiếp đến năng lực xử lý của từng bộ phận và thời gian chờ trung bình của bệnh nhân.

b. **Kích thước hàng đợi (Queue capacity)**  
   - Đại diện cho **số lượng bệnh nhân có thể chờ tối đa** trong từng khu vực.  
   - Quyết định khả năng tiếp nhận bệnh nhân khi hệ thống gần đầy và ảnh hưởng đến tỷ lệ từ chối phục vụ (blocking rate).

### **1.2. Nhiệm vụ nghiên cứu**

Để đạt được mục tiêu của đề tài, dự án thực hiện các nhiệm vụ nghiên cứu sau:
- **Xây dựng mô hình mô phỏng** hệ thống phòng khám nha khoa dựa trên các hàng đợi M/M/c/K, trong đó mỗi tầng hàng đợi biểu diễn một khu chức năng.  
- **Thay đổi và so sánh các kịch bản cơ sở vật chất khác nhau** (số tầng, số server, kích thước hàng chờ) để đánh giá ảnh hưởng của từng yếu tố.
- **Đo lường các chỉ số hiệu năng chính**.
- **So sánh kết quả mô phỏng với công thức lý thuyết** (Little’s Law, M/M/c/K model) để kiểm chứng tính chính xác của mô hình.
- **Phân tích mối tương quan** giữa cơ cấu vật lý của phòng khám và các chỉ số hiệu năng trên, từ đó xác định **các yếu tố cơ sở vật chất có tác động mạnh nhất** đến hiệu quả vận hành của toàn hệ thống.
- **Đưa ra đánh giá định lượng** giúp tối ưu bố cục phòng khám, cải thiện thời gian phục vụ và giảm tình trạng tắc nghẽn;  

### **1.3. Ý nghĩa thực tiễn**

Kết quả nghiên cứu giúp **ban quản lý phòng khám nha khoa**:
- Quy hoạch **số lượng khu chức năng và số ghế khám hợp lý**;  
- Dự báo hiệu năng hệ thống trong các tình huống tải cao;  
- Tối ưu **cơ sở vật chất và quy trình phục vụ** để nâng cao trải nghiệm bệnh nhân và hiệu quả hoạt động của toàn hệ thống.

## **2. Đặc tả hệ thống hàng đợi**

Hệ thống được mô phỏng là một phòng khám nha khoa có nhiều khu vực xử lý bệnh nhân nối tiếp nhau. Mỗi khu vực (Room) trong hệ thống được xem như một hàng đợi (queue) với một hoặc nhiều bác sĩ hoặc thiết bị (servers) phục vụ bệnh nhân theo thứ tự đến. Các hàng đợi này kết nối với nhau thành một **mạng hỗn hợp (hybrid network)** gồm cả tuyến tính (serial) và song song (parallel).

### **2.1. Cấu trúc tổng thể của hệ thống**

Hệ thống gồm **sáu khu vực chính**:

<div align="center">

| **Khu vực** | **Chức năng** |
|:-----------------:|:----------------------:|
| Check-in | Tiếp nhận bệnh nhân và đăng ký khám |
| Treatment Room | Khám và điều trị cơ bản |
| X-Ray Room | Chụp X-quang và kiểm tra hình ảnh |
| Surgery Room | Thực hiện phẫu thuật nha khoa |
| Recovery Room | Theo dõi và chăm sóc sau điều trị |
| Check-out | Thanh toán và hoàn tất thủ tục rời hệ thống |

</div>

### **2.2. Các giả định và giới hạn của hệ thống**

1. **Luồng bệnh nhân liên tục và ngẫu nhiên:**  
   Bệnh nhân đến theo phân phối Poisson; không có đặt lịch trước hoặc nhóm bệnh nhân.

2. **Không có trường hợp bệnh nhân rời khỏi hệ thống trước khi hoàn tất quy trình điều trị:**  
   Sau khi đã vào hàng đợi hoặc đang phục vụ, bệnh nhân sẽ hoàn tất toàn bộ quy trình.

3. **Thời gian di chuyển giữa các khu vực không đáng kể:**  
   Giả định thời gian di chuyển ≈ 0 để tập trung vào chờ và phục vụ.

4. **Các server trong cùng phòng đồng nhất:**  
   Mọi nha sĩ hoặc thiết bị cùng phòng có cùng tốc độ phục vụ trung bình μᵢ.

5. **Phục vụ theo FIFO, không ưu tiên:**  
   Bệnh nhân được phục vụ theo thứ tự đến, không xét mức độ khẩn cấp.

6. **Không có lỗi hoặc downtime:**  
   Các server luôn sẵn sàng trong suốt quá trình mô phỏng.

7. **Hàng đợi có giới hạn (finite K):**  
   Nếu queue đầy, bệnh nhân mới đến bị từ chối phục vụ (blocking).

8. **Phân phối thời gian phục vụ và đến là mũ:**  
   Đảm bảo tính “không nhớ” (memoryless) của hệ thống.

### **2.3. Dịch vụ và kết quả**

Hệ thống phòng khám nha khoa bao gồm các khu vực chức năng, mỗi khu đảm nhận một giai đoạn cụ thể trong quy trình phục vụ bệnh nhân. Dưới đây là mô tả chi tiết các **dịch vụ (Services)** được cung cấp tại từng khu vực và **kết quả (Outcomes)** tương ứng.

**a. Check-in**
- **Service:** Tiếp nhận và đăng ký thông tin bệnh nhân, tạo hồ sơ khám.  
- **Outcome:** Bệnh nhân hoàn tất thủ tục ban đầu và được chuyển sang khu **Treatment Room** để được khám lâm sàng.

**b. Treatment Room**
- **Service:** Bác sĩ tiến hành khám tổng quát, đánh giá tình trạng răng miệng và chỉ định bước tiếp theo.  
- **Outcomes:**  
  1. Bệnh nhân được **chỉ định chụp X-Ray** để kiểm tra chi tiết hơn.  
  2. Bệnh nhân được **chuyển sang Surgery Room** nếu cần phẫu thuật.  
  3. Bệnh nhân **rời hệ thống qua Check-out** nếu không cần điều trị thêm.

**c. X-Ray Room**
- **Service:** Thực hiện chụp X-quang và xử lý hình ảnh răng–hàm–mặt.  
- **Outcome:** Bệnh nhân nhận kết quả chẩn đoán hình ảnh và **trở lại Treatment Room** để được bác sĩ đánh giá kết quả.

**d. Surgery Room**
- **Service:** Thực hiện các thủ thuật hoặc tiểu phẫu nha khoa theo chỉ định.  
- **Outcome:** Sau phẫu thuật, bệnh nhân được **chuyển đến Recovery Room** để theo dõi hồi phục.

**e. Recovery Room**
- **Service:** Theo dõi và chăm sóc bệnh nhân sau điều trị hoặc phẫu thuật, đảm bảo tình trạng ổn định trước khi xuất viện.  
- **Outcomes:**  
  1. Bệnh nhân **rời hệ thống qua Check-out** nếu hồi phục hoàn toàn.  
  2. Bệnh nhân có thể **quay lại Treatment Room** để được kiểm tra thêm nếu có dấu hiệu bất thường.

**f. Check-out**
- **Service:** Thực hiện thanh toán, lưu trữ hồ sơ và cung cấp hướng dẫn sau điều trị.  
- **Outcome:** Bệnh nhân hoàn tất quy trình và **rời khỏi hệ thống**.

### **2.4. Các chỉ số đánh giá hiệu năng**

Để đánh giá hiệu quả hoạt động của hệ thống phòng khám nha khoa được mô phỏng, dự án lựa chọn một tập hợp các **chỉ số hiệu năng (performance metrics)** phản ánh trực tiếp **mức độ đáp ứng và sử dụng tài nguyên** của hệ thống. Các chỉ số này được chọn dựa trên tính chất của mô hình hàng đợi M/M/c/K và mục tiêu của nghiên cứu là **đo lường ảnh hưởng của cơ sở vật chất đến hiệu năng vận hành**.


**a. Thời gian chờ trung bình của bệnh nhân (Average Waiting Time – Wq)**
- **Định nghĩa:** Thời gian trung bình mà bệnh nhân phải chờ trong hàng đợi trước khi được phục vụ.
- **Lý do lựa chọn:** Đây là **chỉ số thể hiện trực tiếp trải nghiệm của bệnh nhân** và phản ánh mức độ tắc nghẽn trong hệ thống. Một hệ thống có **Wq cao** cho thấy cần mở rộng cơ sở vật chất (tăng số server hoặc tăng dung lượng hàng chờ).


**b. Số bệnh nhân trung bình trong hệ thống (Average Number in System – L)**
- **Định nghĩa:** Tổng số bệnh nhân trung bình đang **chờ và được phục vụ** tại một thời điểm.
- **Lý do lựa chọn:** Chỉ số này phản ánh **mức độ tải (workload)** của toàn hệ thống, giúp đánh giá khả năng phục vụ và cân bằng giữa các phòng chức năng. Ngoài ra, chỉ số **L** có thể **đối chiếu với công thức Little’s Law (L = λ × W)** để kiểm chứng tính chính xác của mô phỏng.


**c. Mức độ sử dụng trung bình của server (Server Utilization – ρ)**
- **Định nghĩa:** Tỷ lệ thời gian mà mỗi server (nha sĩ, ghế khám, thiết bị) đang được sử dụng để phục vụ bệnh nhân.
- **Lý do lựa chọn:** Chỉ số **ρ** cho biết **hiệu quả khai thác nguồn lực cơ sở vật chất**. Nếu ρ quá thấp → lãng phí nguồn lực; nếu ρ quá cao → nguy cơ tắc nghẽn và thời gian chờ tăng mạnh. Việc tối ưu ρ giúp đạt **hiệu năng cân bằng** giữa hiệu quả vận hành và chất lượng dịch vụ.


**d. Lưu lượng bệnh nhân được phục vụ (Throughput)**
- **Định nghĩa:** Số lượng bệnh nhân trung bình hoàn tất quy trình phục vụ trong một đơn vị thời gian (ví dụ: bệnh nhân/giờ).
- **Lý do lựa chọn:** Throughput thể hiện **khả năng xử lý tổng thể của hệ thống**, là chỉ số then chốt khi đánh giá tác động của việc tăng server hoặc mở rộng cơ sở vật chất. Hệ thống có throughput cao và Wq thấp thể hiện thiết kế cơ sở vật chất hợp lý.

**e. Thời gian lưu trú trung bình trong hệ thống (Average Time in System – W)**
- **Định nghĩa:** Tổng thời gian trung bình mà bệnh nhân ở lại trong hệ thống, bao gồm thời gian chờ và thời gian phục vụ.  
- **Lý do lựa chọn:** Đây là **thước đo toàn diện** của hiệu năng dịch vụ — từ khi bệnh nhân đến đến khi rời hệ thống. Chỉ số này được sử dụng để kiểm chứng **tính hợp lệ của mô phỏng** thông qua mối quan hệ với các chỉ số khác (Little’s Law: \( L = λW \)).

**f. Xác suất bị từ chối phục vụ (P_block)**
- **Định nghĩa:** Xác suất mà một bệnh nhân bị từ chối phục vụ do hàng đợi đầy (queue capacity K đạt giới hạn).
- **Lý do lựa chọn:** Đây là chỉ số **đặc trưng cho mô hình hàng đợi có dung lượng hữu hạn (M/M/c/K)**, giúp đo lường **mức độ quá tải của hệ thống**. Khi P_block cao, điều đó cho thấy **sức chứa hàng chờ không đáp ứng nhu cầu thực tế**, cần xem xét **mở rộng cơ sở vật chất**.


### **2.5. Các tham số của hệ thống (List of Parameters)**

Mô hình mô phỏng hệ thống phòng khám nha khoa được xây dựng dựa trên chuỗi hàng đợi **M/M/c/K**, trong đó mỗi tầng (queue stage) biểu diễn một khu chức năng cụ thể. Các tham số được chia thành ba nhóm: **tham số cấu trúc**, **tham số vận hành**, và **tham số đo lường**.

**a. Tham số cấu trúc (Structural Parameters)**
<div align="center">

| Ký hiệu | Tên tham số | Ý nghĩa |
|:--------:|:-------------|:--------|
| $S$ | Số tầng hàng đợi (stages) | Số lượng khu chức năng trong hệ thống. |
| $c_i$ | Số server tại tầng $i$ | Mỗi server đại diện cho 1 ghế khám, máy X-ray, hoặc bàn phẫu thuật. |
| $K_i$ | Sức chứa hàng đợi tại tầng $i$ | Tổng số bệnh nhân tối đa (bao gồm cả đang phục vụ và đang chờ). |
| $R$ | Quy tắc điều hướng (routing rule) | Xác suất chuyển tiếp giữa các tầng. |
</div>

**b. Tham số vận hành (Operational Parameters)**
<div align="center">

| Ký hiệu | Tên tham số | Ý nghĩa |
|:--------:|:-------------|:--------|
| $\lambda$ | Tốc độ đến (arrival rate) | Số bệnh nhân trung bình đến hệ thống mỗi đơn vị thời gian. |
| $\mu_i$ | Tốc độ phục vụ tại tầng $i$ (service rate) | Trung bình số bệnh nhân được phục vụ mỗi đơn vị thời gian. |
| $T_{sim}$ | Thời gian mô phỏng (simulation duration) | Tổng thời gian chạy mô phỏng để đạt trạng thái ổn định. |
| $N_{sys}$ | Số bệnh nhân hiện có trong hệ thống | Tổng số bệnh nhân đang tồn tại (đang chờ + đang phục vụ). |
</div>

### **2.6. Các yếu tố cần nghiên cứu**

Trong mô hình mô phỏng, **các tham số được thay đổi có chủ đích** để quan sát ảnh hưởng của chúng đến hiệu năng hệ thống được gọi là **các yếu tố nghiên cứu (factors)**. Mỗi yếu tố có thể được thử nghiệm tại nhiều **mức giá trị khác nhau (levels)**. Những tham số có **tác động mạnh đến hiệu năng tổng thể** của hệ thống sẽ được ưu tiên chọn làm factors, trong khi các tham số ít ảnh hưởng sẽ được **giữ cố định (non-factors)**.

Dưới đây là danh sách các yếu tố (Factors) và mức độ (Levels)

<div align="center">

| Yếu tố (Factor) | Ký hiệu | Giải thích | Các mức độ khảo sát (Levels) | Lý do chọn |
|:--------------------|:-----------:|:----------------|:----------------------------------|:----------------|
| **Số server tại mỗi tầng** | $c_i$ | Số lượng ghế điều trị, máy X-ray hoặc bàn phẫu thuật tại tầng $i$. | 1 / 2 / 3 server per stage | Thể hiện khả năng phục vụ song song, ảnh hưởng mạnh đến throughput và thời gian chờ. |

| **Tốc độ phục vụ tại mỗi tầng** | $\mu_i$ | Tốc độ trung bình mà mỗi server hoàn tất phục vụ một bệnh nhân. | 4 / 6 / 8 bệnh nhân/giờ | Giúp đánh giá tác động của năng suất nhân viên hoặc thiết bị. |

| **Xác suất định tuyến** | $p_{X}, p_{S}, p_{C}$ | ví dụ $\{(0.6,0.2,0.2), (0.4,0.3,0.3)\}$ | Thay đổi tỉ lệ bệnh nhân vào X-Ray hoặc Surgery |

## **3. Phương pháp luận đánh giá**

### **3.1. Lựa chọn phương pháp đánh giá**

Dựa trên **mức độ phức tạp của hệ thống phòng khám nhiều tầng** và **nguồn lực triển khai giới hạn**, đề tài sử dụng kết hợp hai phương pháp chính:

- **Mô hình mô phỏng (Simulation)**  
  - Là phương pháp **chính** để đánh giá hiệu năng.  
  - Cho phép mô tả chi tiết các sự kiện ngẫu nhiên như **arrival**, **service**, **routing**, **blocking**, v.v.  
  - Có thể thu thập được các chỉ số động ($W_q$, $L$, $\rho_i$, $P_{block}$, $X$) mà mô hình toán không phản ánh được đầy đủ.

- **Mô hình phân tích (Analytical Modeling)**  
  - Dùng để **xác minh và hiệu chỉnh kết quả mô phỏng**.  
  - Dựa trên công thức của mô hình hàng đợi M/M/c/K và định luật Little.  
  - Giúp so sánh giữa **giá trị kỳ vọng lý thuyết** và **kết quả thực nghiệm mô phỏng** để đánh giá sai số.

### **3.2. Mô hình hóa hệ thống**

Phần này mô tả cách **các thực thể của phòng khám** được **ánh xạ (mapped)** thành các thành phần trong mô hình hàng đợi M/M/c/K dùng cho mô phỏng. Dưới đây là các nguyên tắc ánh xạ:

- **Mỗi khu chức năng thực tế** (ví dụ: Check-in, Treatment, X-Ray, Surgery, Recovery, Check-out) → **một hàng đợi** trong mô hình.  
- **Phòng tại khu chức năng / quầy phục vụ** → **server** trong hàng đợi.  
- **Số chỗ ngồi chờ / diện tích chờ** → **kích thước hàng đợi**.  
- **Bệnh nhân đến ngẫu nhiên** → mô phỏng theo **phân phối Poisson**, trong đó khoảng **thời gian giữa hai lượt đến liên tiếp** tuân theo **phân phối mũ** với tham số λ.  
- **Thời gian phục vụ** (thời gian khám, thời gian phẫu thuật, thời gian chụp X-quang, v.v.) → được lấy ngẫu nhiên từ **phân phối mũ** với tham số μ.  
- **Luồng điều hướng bệnh nhân giữa phòng** → mô tả bằng **quy tắc định tuyến (routing)**, trong đó mỗi bệnh nhân có xác suất nhất định để được chuyển tiếp sang các hàng đợi.

Dưới đây là bảng ánh xạ chi tiết:

<div align="center">

| Thực thể / Hiện tượng thực tế | Định nghĩa trong mô hình | Ghi chú / Ví dụ |
|:------------------------------|:-------------------------|:-----------------|
| Bệnh nhân đến phòng khám | Arrival process | Ví dụ: 5 bệnh nhân/giờ → interarrival ~ Exp(1/5 giờ) |
| Phòng tại khu chức năng / quầy phục vụ | Server | Treatment: c₂ = 3 (3 phòng khám) |
| Ghế/diện tích khu chờ | Queue capacity | Reception K₁ = 5, Treatment K₂ = 10 |
| Thời gian đăng ký / khám / X-Ray / phẫu thuật / thanh toán | Service time | μ₂ = 8 bệnh nhân/giờ → mean service = 1/8 giờ |
| Bệnh nhân bị từ chối khi chờ đầy | Blocking | Nếu queue đầy, arrival bị drop |
| Luồng tuần tự / song song giữa các khu | Network topology | Treatment → (X-Ray, Surgery) → Recovery |

</div>

### **3.4. Thiết kế thí nghiệm**

Dựa trên **danh sách các yếu tố (factors)** và **mức thay đổi (levels)** đã định nghĩa ở Mục 2.6, giai đoạn này xác định **tập hợp các kịch bản mô phỏng (simulation experiments)** nhằm:

- **Thu thập thông tin tối đa** về mối quan hệ giữa các yếu tố cơ sở vật chất và các chỉ số hiệu năng.  
- **Giảm thiểu số lần mô phỏng cần thiết**, đảm bảo hiệu quả tính toán và khả năng mở rộng.

**a. Mục tiêu thiết kế thí nghiệm**
- Khám phá **tác động riêng và kết hợp** giữa các yếu tố ($c_i$, $K_i$) trong các điều kiện khác nhau ($\mu_i$, $p_X$, $p_S$) đến các chỉ số hiệu năng ($W_q$, $L$, $\rho$, $X$, $P_{block}$).  
- So sánh kết quả giữa **3 loại tải công việc (workloads)** khác nhau đã chọn:  
  - A – Normal Load  
  - B – Peak Load  
  - C – Surgery Intensive 

**b. Phương pháp thiết kế**

Để cân bằng giữa **độ chính xác** và **chi phí mô phỏng**, hệ thống thí nghiệm được xây dựng theo **chiến lược hai giai đoạn (Two-Phase Approach):**

| **Giai đoạn** | **Mục tiêu** | **Đặc điểm thiết kế** | **Kết quả mong đợi** |
|:---------------|:-------------|:-----------------------|:----------------------|
| **Phase 1 – Khảo sát tổng quan** | Xác định yếu tố ảnh hưởng mạnh nhất | Nhiều yếu tố (5 factors), mỗi yếu tố chỉ có **2 mức** (thấp/cao) | Phát hiện yếu tố quan trọng, loại bỏ yếu tố ít tác động |
| **Phase 2 – Phân tích chi tiết** | Nghiên cứu sâu các yếu tố quan trọng | Giữ lại **2 yếu tố chính**, tăng số mức **(3 levels)** | Xác định mối quan hệ phi tuyến và mức tối ưu hiệu năng |


**c. Cấu hình thí nghiệm**

Để cân bằng giữa độ bao phủ và chi phí mô phỏng, hệ thống được thiết kế gồm **hai giai đoạn**:


**Phase 1 – Khảo sát tổng quan**

**Mục tiêu:** Giai đoạn này nhằm **xác định các yếu tố cơ sở vật chất có ảnh hưởng lớn nhất** đến hiệu năng của hệ thống phòng khám nha khoa. Để đơn giản hóa mô hình mà vẫn đảm bảo tính đại diện cho quy trình thực tế, hệ thống được chia thành **ba cụm chức năng chính** tương ứng với các giai đoạn hoạt động độc lập:

<div align="center">

| **Cụm chức năng** | **Bao gồm** | **Đặc trưng hoạt động** |
|:------------------|:-------------|:--------------------------|
| **Cụm Thủ tục** | Check-in + Check-out | Tiếp nhận, lưu trữ và thanh toán hồ sơ bệnh nhân |
| **Cụm Chẩn đoán** | Treatment + X-ray | Khám tổng quát, chẩn đoán hình ảnh và lập phác đồ điều trị |
| **Cụm Phẫu thuật** | Surgery + Recovery | Thực hiện phẫu thuật, theo dõi và chăm sóc sau điều trị |
</div>

**Workloads sử dụng:**

<div align="center">

| **Loại tải** | **Mô tả** |
|:-------------|:-----------|
| **A – Normal Load** | Hoạt động trung bình, phân bổ bệnh nhân ổn định |
| **B – Peak Load** | Giờ cao điểm, lưu lượng bệnh nhân tăng cao |
</div>

**Lý do chọn mô hình 3 cụm:**

Nếu xét riêng từng phòng, số lượng kịch bản mô phỏng sẽ tăng theo cấp số nhân. Cụ thể, với 6 tầng hàng đợi (Check-in, Treatment, X-Ray, Surgery, Recovery, Check-out), mỗi tầng có hai yếu tố cơ sở vật chất (số server, kích thước hàng đợi), tổng số tổ hợp cấu hình là:

$$
2^{12} = 4096
$$

Khi xét thêm hai loại tải (Normal và Peak), tổng số kịch bản đạt **8192**, và nếu mỗi cấu hình lặp lại 3 lần, tổng số lượt mô phỏng vượt **24 000**.

Do đó, các phòng có chức năng tương đồng được gom lại thành ba **cụm chức năng đại diện**.  
Mỗi cụm chia sẻ cùng cấu hình hàng đợi $(c_i, K_i)$, giúp:
- **Giảm tổng số yếu tố khảo sát từ 12 xuống 6**,  
- **Giảm tổng số kịch bản từ 8192 xuống còn 128**,  
- Vẫn đảm bảo phản ánh được các giai đoạn chính của quy trình phục vụ.

**Cấu hình mô phỏng cho từng cụm:**

|:-----------|:----------------|:-------------------------:|:------------------------:|:------------:|
| **Thủ tục** | Số server ($c_p$) | 1 | 3 | server |

| **Chẩn đoán** | Số server ($c_d$) | 3 | 9 | server |

$$

$$

Mỗi kịch bản mô phỏng chạy trong **3600 đơn vị thời gian**, **lặp lại 5 lần (replications)** để ước lượng trung bình và phương sai ổn định.

- Phát hiện **mối tương quan** giữa các yếu tố cơ sở vật chất và hiệu năng hệ thống.  
- Tạo cơ sở cho **phân tích thống kê (ANOVA / Regression)** ở phần đánh giá sau.  
- Xây dựng dữ liệu đủ phong phú để so sánh giữa **mô phỏng và mô hình lý thuyết (M/M/c/K)**.

## **4. Triển khai mã nguồn**

Để kiểm chứng các mục tiêu và phương pháp luận đã đề ra, chúng tôi xây dựng một mô hình mô phỏng sự kiện rời rạc (Discrete-Event Simulation) bằng Python và thư viện **simpy**. Phần này trình bày mã nguồn chính được sử dụng để định nghĩa các thành phần của hệ thống.

```python
import simpy
import random
import numpy as np

# --- 1. Định nghĩa các Tham số Cố định (dựa trên Bảng 2 & 3) ---

# Giả định đơn vị thời gian cơ sở là PHÚT.

# Service Times (1/mu) - Thời gian phục vụ trung bình (phút)
# Dựa trên Bảng 2: Mô hình kendall [cite: 193]
# Ví dụ: mu = 1/4 (bệnh nhân/phút) => service_time = 4 phút
SERVICE_TIMES = {
    'Check-in': 4.0,     # 1 / (1/4)
    'Treatment': 10.0,   # 1 / (1/10)
    'X-Ray': 7.0,        # 1 / (1/7)
    'Surgery': 15.0,     # 1 / (1/15)
    'Recovery': 20.0,    # 1 / (1/20)
    'Check-out': 3.0,    # 1 / (1/3)
}

# Routing Probabilities - Xác suất rẽ nhánh
# Dựa trên Bảng 3: Ma trận xác suất P_ij 
ROUTING_PROBS = {
    'Treatment': {
        'destinations': ['X-Ray', 'Surgery', 'Check-out'],
        'probabilities': [0.3, 0.2, 0.5]
    },
    'Recovery': {
        'destinations': ['Treatment', 'Check-out'],
        'probabilities': [0.1, 0.9]
    }
    # Các luồng khác là cố định (tuyến tính)
}

print("Các tham số cố định đã được định nghĩa.")
```

### **4.1 Định nghĩa khu vực hàng đợi (Clinic stage)**
Chúng tôi hiện thực mỗi khu vực chức năng (Check-in, Treatment, v.v.)  như một lớp ClinicStage độc lập. Lớp này mô phỏng một hệ thống hàng đợi **M/M/c/K**.
- **c** (servers): Số lượng bác sĩ/ghế/máy, được mô phỏng bằng **simpy.Resource**.
- **K** (capacity): Sức chứa tối đa của hệ thống (chờ + phục vụ).
- Logic **Blocking**: Nếu một bệnh nhân đến khi hệ thống đã đầy (số đang chờ + số đang phục vụ >= K), bệnh nhân đó sẽ bị từ chối (blocked) và rời đi.

```python
class ClinicStage:
    """
    Đại diện cho một khu vực chức năng (M/M/c/K).
    Bao gồm các server (c) và giới hạn sức chứa (K).
    """
    def __init__(self, env, name, c, K, mu, metrics_collector):
        self.env = env
        self.name = name
        self.K = K  # Tổng sức chứa (chờ + phục vụ)
        self.mu = mu # Tốc độ phục vụ (service rate)
        self.metrics = metrics_collector
        
        # 'c' servers (bác sĩ, máy X-quang, ...)
        self.servers = simpy.Resource(env, capacity=c)
        
        # THÊM: Flag để biết có phải Check-in không (để quyết định block hay chờ)
        self.is_checkin = (name == 'Check-in')
        
        # THÊM: Lưu c để tính ρ
        self.c = c

    def num_in_system(self):
        """Trả về tổng số bệnh nhân (đang chờ + đang phục vụ)"""
        return len(self.servers.queue) + len(self.servers.users)
    
    # THÊM: Method để lấy số bệnh nhân đang chờ (Lq)
    def num_in_queue(self):
        return len(self.servers.queue)
    
    # THÊM: Method để lấy số server đang bận (cho ρ = num_busy / c)
    def num_busy_servers(self):
        return len(self.servers.users)

    def handle_patient(self, patient_id):
        """
        Xử lý quy trình của 1 bệnh nhân khi đến khu vực này.
        Trả về True nếu được phục vụ, False nếu bị block.
        """
        t_arrival = self.env.now
        
        # THAY ĐỔI: Nếu là Check-in và đầy, block ngay (return False)
        if self.is_checkin and self.num_in_system() >= self.K:
            self.metrics.add_block(self.name, patient_id)
            return False  # Bị block chỉ ở Check-in
        
        # THÊM: Nếu không phải Check-in và đầy, chờ đến khi có chỗ (vòng lặp poll)
        while not self.is_checkin and self.num_in_system() >= self.K:
            # Chờ một khoảng thời gian nhỏ (ví dụ: 0.1 phút) rồi kiểm tra lại
            # Điều này mô phỏng "waiting area" - có thể điều chỉnh timeout để tránh CPU heavy
            yield self.env.timeout(0.1)  # Timeout nhỏ để poll, tránh loop chặt
            
            # THÊM (TÙY CHỌN): Ghi nhận thời gian chờ thêm này vào metrics (nếu cần, để tính vào Wq)
            # extra_wait = self.env.now - t_wait_start  # Có thể add nếu định nghĩa t_wait_start trước while

        # Bây giờ request server (đã chắc chắn có chỗ)
        with self.servers.request() as req:
            yield req
            
            # 3. Tính toán thời gian chờ (Wq)
            t_start_service = self.env.now
            wait_time = t_start_service - t_arrival
            self.metrics.add_wait_time(self.name, wait_time)
            
            # 4. Quá trình phục vụ (Service Time)
            # Thời gian phục vụ tuân theo phân phối mũ [cite: 76]
            service_time = random.expovariate(self.mu)
            yield self.env.timeout(service_time)
        
        # Bệnh nhân đã được phục vụ
        return True
```

### **4.2. Định nghĩa Luồng Quy trình (Patient Process)**
Chúng tôi định nghĩa một "process" (tiến trình) cho mỗi bệnh nhân. Tiến trình này mô phỏng toàn bộ hành trình của bệnh nhân, từ **Check-in** đến **Check-out**, dựa trên sơ đồ Hình 1 và ma trận xác suất Bảng 3.

```python
class ClinicSimulation:
    """
    Lớp điều khiển chính, chứa tất cả các khu vực (stages)
    và logic điều hướng (routing).
    """
    def __init__(self, env, config, metrics_collector):
        self.env = env
        self.metrics = metrics_collector
        self.stages = {}
        
        # Khởi tạo tất cả các khu vực chức năng
        for stage_name, mu in SERVICE_TIMES.items():
            stage_config = config.get(stage_name, {})
            c = stage_config.get('c', 1) # Mặc định là 1 server
            K = stage_config.get('K', 10) # Mặc định K=10
            
            self.stages[stage_name] = ClinicStage(
                env, stage_name, c, K, (1.0/mu), metrics_collector
            )
        
        # THÊM: Khởi động process sampler để thu thập samples cho L, Lq, ρ
        env.process(self.sampler())

    def get_next_stage_name(self, current_stage_name):
        """
        Quyết định khu vực tiếp theo dựa trên logic điều hướng.
        """
        if current_stage_name == 'Check-in':
            return 'Treatment'
        
        if current_stage_name == 'X-Ray':
            return 'Treatment' # [cite: 94]
            
        if current_stage_name == 'Surgery':
            return 'Recovery' # [cite: 98]
        
        if current_stage_name == 'Treatment':
            # Rẽ nhánh ngẫu nhiên 
            dest = ROUTING_PROBS['Treatment']['destinations']
            probs = ROUTING_PROBS['Treatment']['probabilities']
            return random.choices(dest, probs)[0]
            
        if current_stage_name == 'Recovery':
            # Rẽ nhánh ngẫu nhiên 
            dest = ROUTING_PROBS['Recovery']['destinations']
            probs = ROUTING_PROBS['Recovery']['probabilities']
            return random.choices(dest, probs)[0]
            
        if current_stage_name == 'Check-out':
            return 'DEPART' # Rời hệ thống

        return 'DEPART'
    
    # THÊM: Process sampler để định kỳ thu thập dữ liệu
    def sampler(self):
        while True:
            # Thu thập samples cho mỗi stage
            for stage_name, stage in self.stages.items():
                self.metrics.add_sample(
                    stage_name,
                    stage.num_in_system(),  # L_sample
                    stage.num_in_queue(),   # Lq_sample
                    stage.num_busy_servers() / stage.c if stage.c > 0 else 0  # rho_sample
                )
            
            # Định kỳ mỗi 1 phút (có thể điều chỉnh)
            yield self.env.timeout(1.0)
```

Bổ sung 2 hàm chức năng: **patient_process** và **source**. 

```python
def patient_process(env, patient_id, clinic, metrics_collector):
    """
    Mô phỏng quy trình của một bệnh nhân duy nhất.
    """
    t_system_arrival = env.now
    current_stage_name = 'Check-in'
    
    while current_stage_name != 'DEPART':
        stage = clinic.stages[current_stage_name]
        
        # Xử lý tại khu vực hiện tại
        was_served = yield env.process(stage.handle_patient(patient_id))
        
        # THAY ĐỔI NHỎ: Chỉ DEPART nếu không served (chỉ xảy ra ở Check-in)
        if not was_served:
            current_stage_name = 'DEPART'
        else:
            current_stage_name = clinic.get_next_stage_name(stage.name)

    # Ghi nhận khi bệnh nhân rời hệ thống
    t_system_departure = env.now
    total_system_time = t_system_departure - t_system_arrival
    metrics_collector.add_system_time(total_system_time, patient_id)

def source(env, lambda_rate, clinic, metrics_collector):
    """
    Nguồn phát sinh bệnh nhân (Arrival Process) theo phân phối Poisson.
    """
    patient_id = 0
    while True:
        # Thời gian giữa hai lượt đến tuân theo phân phối mũ [cite: 184]
        interarrival_time = random.expovariate(lambda_rate)
        yield env.timeout(interarrival_time)
        
        patient_id += 1
        metrics_collector.add_arrival(patient_id)
        # Tạo một tiến trình bệnh nhân mới
        env.process(patient_process(env, patient_id, clinic, metrics_collector))
```

### **4.3. Thu thập Chỉ số hiệu năng hệ thống**
Để đo lường hiệu năng, chúng tôi tạo một lớp **MetricsCollector** để ghi lại các sự kiện (đến, chờ, block, hoàn tất) và tính toán các chỉ số cuối cùng như $W_q$, $P_{\text{block}}$, và Throughput.

```python
class MetricsCollector:
    """Thu thập và tính toán các chỉ số hiệu năng."""
    def __init__(self):
        self.wait_times = {} # {stage: [w1, w2, ...]}
        self.system_times = []
        self.arrivals = 0
        self.blocked_at = {} # {stage: count}
        self.completed = 0
        self.unique_patients = set()
        self.blocked_patients = set()
        
        # THÊM: Storage cho samples per stage để tính L, Lq, ρ
        self.samples = {}  # {stage: {'L_samples': [], 'Lq_samples': [], 'rho_samples': []}}

    def add_arrival(self, patient_id):
        self.arrivals += 1
        self.unique_patients.add(patient_id)

    def add_wait_time(self, stage_name, wait_time):
        if stage_name not in self.wait_times:
            self.wait_times[stage_name] = []
        self.wait_times[stage_name].append(wait_time)

    def add_system_time(self, system_time, patient_id):
        if patient_id not in self.blocked_patients:
            self.system_times.append(system_time)
            self.completed += 1

    def add_block(self, stage_name, patient_id):
        if stage_name not in self.blocked_at:
            self.blocked_at[stage_name] = 0
        self.blocked_at[stage_name] += 1
        self.blocked_patients.add(patient_id)
    
    # THÊM: Hàm để add sample từ sampler
    def add_sample(self, stage_name, L_sample, Lq_sample, rho_sample):
        if stage_name not in self.samples:
            self.samples[stage_name] = {'L_samples': [], 'Lq_samples': [], 'rho_samples': []}
        self.samples[stage_name]['L_samples'].append(L_sample)
        self.samples[stage_name]['Lq_samples'].append(Lq_sample)
        self.samples[stage_name]['rho_samples'].append(rho_sample)

    def calculate_results(self, sim_time, lambda_rate_per_min):
        total_arrivals = len(self.unique_patients)
        total_blocked = len(self.blocked_patients)
        
        p_block = total_blocked / total_arrivals if total_arrivals > 0 else 0
        
        all_waits = [t for waits in self.wait_times.values() for t in waits]
        avg_wait_time = np.mean(all_waits) if all_waits else 0
        
        avg_system_time = np.mean(self.system_times) if self.system_times else 0
        
        throughput_per_min = self.completed / sim_time
        throughput_per_hour = throughput_per_min * 60
        
        # THÊM: Tính L, Lq, ρ, P_idle per stage và tổng
        per_stage_metrics = {}
        total_L = 0
        total_Lq = 0
        total_rho = 0
        num_stages = len(self.samples)
        
        for stage_name, data in self.samples.items():
            L_avg = np.mean(data['L_samples']) if data['L_samples'] else 0
            Lq_avg = np.mean(data['Lq_samples']) if data['Lq_samples'] else 0
            rho_avg = np.mean(data['rho_samples']) if data['rho_samples'] else 0
            p_idle = 1 - rho_avg
            
            per_stage_metrics[stage_name] = {
                'L (Avg in System)': L_avg,
                'Lq (Avg in Queue)': Lq_avg,
                'p (Utilization)': rho_avg,
                'P_idle': p_idle
            }
            
            total_L += L_avg
            total_Lq += Lq_avg
            total_rho += rho_avg
        
        avg_rho = total_rho / num_stages if num_stages > 0 else 0
        avg_p_idle = 1 - avg_rho
        
        # THÊM: Verify Little’s Law: estimated_L = λ * W (λ per min * avg_system_time)
        lambda_rate_per_min_effective = lambda_rate_per_min * (1 - p_block)  # Adjust cho block
        estimated_L = lambda_rate_per_min_effective * avg_system_time
        littles_law_diff = abs(estimated_L - total_L) / total_L if total_L > 0 else 0
        
        results = {
            'Wq (Avg Wait Time)': avg_wait_time,
            'W (Avg System Time)': avg_system_time,
            'P_block (Probability)': p_block,
            'Throughput (patients/hour)': throughput_per_hour,
            'Total Arrivals': total_arrivals,
            'Total Completed': self.completed,
            'Total Blocked': total_blocked,
            
            # THÊM: Tổng hệ thống
            'L (Avg Number in System - Total)': total_L,
            'Lq (Avg Number in Queue - Total)': total_Lq,
            'ρ (Avg Utilization - Total)': avg_rho,
            'P_idle (Avg - Total)': avg_p_idle,
            
            # THÊM: Per stage
            'Per Stage Metrics': per_stage_metrics,
            
            # THÊM: Verify
            'Little’s Law Estimated L': estimated_L,
            'Little’s Law Difference (%)': littles_law_diff * 100
        }
        
        return results
```

### **4.4. Cấu hình tham số và chạy mô phỏng hệ thống**

Dưới đây là một ví dụ chạy thí nghiệm, tương ứng với một kịch bản trong "Phase 2" 11 (2 server, sức chứa 6) và "Workload A - Normal Load" (ví dụ: 5 bệnh nhân/giờ).

```python
# --- 4.4. Chạy kịch bản thí nghiệm ---

# Thiết lập kịch bản (ví dụ từ Phase 2, Level 2) [cite: 228, 232]
# Giả định 'c' và 'K' đồng nhất cho tất cả các tầng
C_SERVERS = 2  # 2 servers
K_CAPACITY = 6 # Sức chứa 6

# Workload (ví dụ: 5 bệnh nhân/giờ) [cite: 161]
# Đổi sang đơn vị PHÚT
LAMBDA_PER_HOUR = 5.0
LAMBDA_PER_MINUTE = LAMBDA_PER_HOUR / 60.0

# Thời gian mô phỏng (ví dụ: 1 ngày làm việc 8 tiếng)
# "3600 đơn vị thời gian"[cite: 227], nếu là phút thì là 60 giờ.
# Chúng ta sẽ chạy 480 phút (8 giờ) để mô phỏng 1 ngày.
SIM_TIME = 8 * 60 

# Xây dựng cấu hình
experiment_config = {}
for stage_name in SERVICE_TIMES.keys():
    experiment_config[stage_name] = {'c': C_SERVERS, 'K': K_CAPACITY}

# Khởi tạo môi trường
env = simpy.Environment()
metrics = MetricsCollector()
clinic = ClinicSimulation(env, experiment_config, metrics)

# Chạy mô phỏng
env.process(source(env, LAMBDA_PER_MINUTE, clinic, metrics))
env.run(until=SIM_TIME)

# Tính toán và In kết quả
# THÊM: Truyền lambda_rate_per_min vào calculate để verify Little’s Law
results = metrics.calculate_results(sim_time=SIM_TIME, lambda_rate_per_min=LAMBDA_PER_MINUTE)

print("--- KẾT QUẢ MÔ PHỎNG ---")
print(f"Cấu hình: c={C_SERVERS}, K={K_CAPACITY}, lambda={LAMBDA_PER_HOUR}/giờ")
print(f"Thời gian mô phỏng: {SIM_TIME} phút")
print("--------------------------")
for key, value in results.items():
    if key == 'Per Stage Metrics':
        print(f"{key}:")
        for stage, metrics_dict in value.items():
            print(f"  {stage}:")
            for m_key, m_value in metrics_dict.items():
                print(f"    {m_key}: {m_value:.4f}")
    else:
        print(f"{key}: {value:.4f}")
```
