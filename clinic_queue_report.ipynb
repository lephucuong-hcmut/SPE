## 1. Mục tiêu 

Mục tiêu của bài tập lớn này là xây dựng và mô phỏng một mô hình phòng khám y tế nhiều giai đoạn (multi-stage healthcare queuing system) nhằm phân tích và đánh giá hiệu năng hoạt động của hệ thống trong các điều kiện khác nhau. Hệ thống được thiết kế để mô phỏng quy trình bệnh nhân từ lúc đăng ký khám bệnh (Check-in) đến khi hoàn tất điều trị và rời phòng khám (Check-out), đi qua nhiều phòng xử lý trung gian như phòng khám (Treatment Room), phòng chụp X-quang (X-ray Room), phòng phẫu thuật (Surgery Room), và phòng hồi phục (Recovery Room). Kết quả mô phỏng sẽ giúp đưa ra những khuyến nghị cho việc tối ưu hoá vận hành phòng khám. Mục tiêu của mô hình mô phỏng là đánh giá hiệu năng hoạt động của hệ thống phòng khám và phân tích ảnh hưởng của các yếu tố đến chất lượng dịch vụ. 

Cụ thể, mô hình tập trung vào các chỉ tiêu sau:

**1. Thời gian chờ trung bình của bệnh nhân (Average Waiting Time)**

**2. Mức sử dụng tài nguyên của từng phòng (Server Utilization)**

**3. Lưu lượng bệnh nhân được phục vụ mỗi giờ (Throughput)**

**4. Số lượng bác sĩ hoặc thiết bị tại mỗi phòng (Number of Servers)**

**5. Lưu lượng bệnh nhân đến theo thời gian (Patient Arrival Rate)**


Từ các chỉ số trên, mô hình giúp xác định các nút thắt (bottlenecks) trong quy trình khám bệnh và đề xuất hướng tối ưu hóa về phân bổ nhân lực hoặc thiết bị y tế.

## 2. Mô tả hệ thống 

Hệ thống được mô phỏng là một phòng khám đa khoa có nhiều khu vực xử lý bệnh nhân nối tiếp nhau. Mỗi khu vực (Room) trong hệ thống được xem như một hàng đợi (queue) với một hoặc nhiều bác sĩ hoặc thiết bị (servers) phục vụ bệnh nhân theo thứ tự đến.

# Hình 1: Sơ đồ mô tả của hệ thống

**Quy trình hoạt động tổng quát của bệnh nhân được mô tả như sau:**

**Check-in:** Bệnh nhân đến phòng khám và đăng ký thông tin cá nhân. Sau khi hoàn tất, họ sẽ được chuyển đến khu Treatment Room để gặp bác sĩ khám tổng quát.

**Treatment Room:** Tại đây, bác sĩ đánh giá tình trạng bệnh nhân và quyết định bước điều trị tiếp theo. Sau khi được khám, bệnh nhân có thể:

1. Được chỉ định chụp X-ray (sau đó quay lại Treatment Room)

2. Được chỉ định phẫu thuật nhỏ (Surgery Room) và sau đó nghỉ tại Recovery Room

3. Hoặc rời hệ thống nếu không cần điều trị thêm (đi thẳng đến Check-out).

**X-ray Room:** Bệnh nhân được chụp X-quang và trả kết quả, sau đó quay lại Treatment Room để được đọc kết quả 

**Surgery Room:** Bệnh nhân được thực hiện tiểu phẫu, sau đó được chuyển đến Recovery Room để theo dõi.

**Recovery Room:** Bệnh nhân hồi phục trong một khoảng thời gian nhất định trước khi rời hệ thống hoặc quay lại Treatment Room để nếu có sự cố xảy ra.

**Check-out:** Bệnh nhân hoàn tất các thủ tục và rời phòng khám.

## 3. Mô hình hoá hệ thống

**Mỗi khu vực được mô hình hóa là một queue. Mỗi queue có:**
1. Phân phối đến (arrival distribution)
2. Phân phối phục vụ (service distribution)
3. Số server (bác sĩ hoặc thiết bị)

**Dạng biểu diễn theo Kendall notation:** **A/S/c**

Trong đó: 

- **A:** kiểu phân phối thời gian đến (Arrival)

- **S:** kiểu phân phối thời gian phục vụ (Service)

- **c:** số server

Từ đây, ta giả định Kendall cho từng phòng và xác suất rẽ nhánh giữa các phòng (Routing Probabilities) trong hệ thống như sau:

| **Phòng** | **Số server c (bác sĩ/thiết bị)** | **Ký hiệu Kendall** | **Service rate μ (phút)** | **Ghi chú** |
|:-----------|:------------------------------:|:-------------------:|:------------------------------------------:|:------------|
| Check-in | 1 | M/M/1 | μ₁ = 1 / 4  | Lễ tân đăng ký, trung bình 4 phút/bệnh nhân |
| Treatment Room | 3  | M/M/3 | μ₂ = 1 / 10  | Bác sĩ khám tổng quát, 3 người |
| X-ray Room | 2  | M/M/2 | μ₃ = 1 / 7 | Máy chụp X-quang, trung bình 7.5 phút |
| Surgery Room | 1 | M/M/1 | μ₄ = 1 / 15  | Phẫu thuật nhỏ, trung bình 15 phút |
| Recovery Room | 2  | M/M/2 | μ₅ = 1 / 20  | Giường hồi sức |
| Check-out | 1 | M/M/1 | μ₆ = 1 / 3  | Thanh toán và rời hệ thống |

# Bảng 2: Ma trận xác suất rẽ nhánh ($P_{ij}$) của bệnh nhân
Toàn bộ hệ thống có thể được mô hình hóa như một mạng hàng đợi mở (open queueing network) — nơi bệnh nhân đi vào từ bên ngoài (Check-in), được phục vụ qua nhiều trạm (phòng), rồi rời hệ thống qua Check-out. 

**Mỗi node i được đặc trưng bởi:**

- $\lambda_i$: tốc độ đến *(arrival rate)* của node *i*  
- $\mu_i$: tốc độ phục vụ trung bình *(service rate)*  
- $c_i$: số server tại node *i*  
- $P_{ij}$: xác suất khách từ node *i* chuyển sang node *j*

**Cân bằng lưu lượng (Traffic Equations) và điều kiện ổn định của mô hình**

Tốc độ đến tổng tại mỗi node (ngoại trừ Check-in) được xác định bởi phương trình cân bằng:
$$
\lambda_i = \sum_j \lambda_j P_{ji}
$$
Tốc độ đến trung bình (arrival rate) phải nhỏ hơn tốc độ phục vụ trung bình của hệ thống: 
$$
\lambda_i < c_i \times \mu_i
$$

**Khi đó, ta giả định tốc độ vào tại quầy Check in ($\lambda_1$) là một số cho trước ta sẽ suy ra tốc độ vào và độ ổn định của các node còn lại.**
